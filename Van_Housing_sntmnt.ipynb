### Canada Housing Market Tweets - Sentiment Analysis

```python
import pandas as pd
import numpy as np
```

```python
t = pd.read_csv("twitterData.csv")
print(t.value_counts('sentiment'))
```

```python
t
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_df=0.95,min_df=2,stop_words="english")
#convert the vectorizer.fit_transform() into a data frame so we can perform functions on it. 
dfTFIDF = pd.DataFrame.sparse.from_spmatrix(vectorizer.fit_transform(t['tweet'])) 
column_name_new = vectorizer.get_feature_names_out()
dfTFIDF.columns = column_name_new
dfTFIDF
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
X = dfTFIDF
Y = t['sentiment']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.2, random_state=42)
```

```python
MnB = MultinomialNB()
MnB.fit(X_train,Y_train)
MnB.score(X_train,Y_train)
```

```python
y_pred = MnB.predict(X_test)
y_pred
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
con_matrix=confusion_matrix(Y_test,y_pred)
sns.heatmap(con_matrix, annot=True)
```

```python
X_new = ["vancouver housing price is too high","canada\'s housing market had the best year ever with price growth continues across canada"]
```

```python
X_new_trans = pd.DataFrame.sparse.from_spmatrix(vectorizer.transform(X_new))
y_pred_new = MnB.predict(X_new_trans)
print(y_pred_new)
```
